```python
#Import Library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Load the csv data

df=pd.read_csv('airbnb.csv')
```

```python
df.head()
```

**PIvot**

```python
df_slice = df[['person_capacity', 'room_type', 'Unnamed: 0']]
slice_pivot = df_slice.pivot_table(index=['person_capacity'], columns=['room_type'], \
                                        aggfunc=np.size,fill_value=0)
```

```python
slice_pivot
```

```python
df_slice = df[['person_capacity', 'room_type', 'bedrooms', 'Unnamed: 0']]
slice_pivot = df_slice.pivot_table(index=['bedrooms', 'person_capacity'], columns=['room_type'], \
                                        aggfunc=np.size,fill_value=0)
```

```python
slice_pivot
```

**Merge, Join, Concatenate and Compare**

**Concat**

```python
df1 = df[['Unnamed: 0', 'room_type', 'person_capacity']][0:5]
```

```python
df1
```

```python
df2 = df[['Unnamed: 0', 'room_type', 'person_capacity']][5:10]
```

**Merge**

```python
df3 = df[['Unnamed: 0', 'room_type', 'person_capacity']][0:5]
```

```python
df3
```

```python
df4 = df[['Unnamed: 0', 'bedrooms', 'guest_satisfaction']][0:5]
```

```python
df4
```

```python
df_merge = pd.merge(df3, df4, how="left", on=["Unnamed: 0"])
df_merge
```
